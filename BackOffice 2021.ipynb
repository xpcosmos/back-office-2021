# Questão I:

```python
soma = []

a = [] # Quinze
b = [] # Vinte e oito
c = [] # Trinta e sete
d = [] # Quarenta e dois
e = [] # Sessenta

freq_a = []
freq_b = []
freq_c = []
freq_d = []
freq_e = []


for m in range(15):
  for i in range(10000):
    x = random.randint(0, 60)
    y = random.randint(0, 60)
    z = random.randint(0, 60)
    if x+y+z == 15:
      a.append(1)
    elif x+y+z == 28:
      b.append(1)
    elif x+y+z == 37:
      c.append(1)
    elif x+y+z == 42:
      d.append(1)
    elif x+y+z == 60:
      e.append(1)
  freq_a.append(sum(a))
  freq_b.append(sum(b))
  freq_c.append(sum(c))
  freq_d.append(sum(d))
  freq_e.append(sum(e))

freq = {'15': sum(freq_a), '28':sum(freq_b), '37':sum(freq_c), '42': sum(freq_d), '60':sum(freq_e)}

freq_df = pd.DataFrame(freq, index= [0])

total = freq_df.sum(axis=1)

for i in freq_df:
  freq_df[i] = freq_df[i]  /total
  freq_df[i]  = freq_df[i]  * 100

print('Frequência de aparecimento com base em 10.000 número aleatório gerados 15 vezes:\n')
freq_df
```

```python
from math import sqrt
```

```python
statement_1 = []
statement_2 = []
both_statement = []
num_tentativas= []

for x in range(100000000):
  if sqrt(x) == 36:
    statement_1.append(x)
  elif x*(x-6) ==0:
    statement_2.append(x)
  elif sqrt(x) == 36 and x*(x-6) == 0:
    both_statement.append(x)
  num_tentativas.append(1)

print(f'Statement 1 is:\t{statement_1}\n\nStatement 2 is:\t {statement_2}\n\nBoth Statements is:\t {both_statement}\n\nNúmero de tentativas realizadas:\t{len(num_tentativas)}')
```
